```python
#Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Holidays Data

```python
#Read Dataset and Get Inital few rows and Dataset dataType information
holidaysData = pd.read_csv('holidays_events.csv')

holidaysData.head(), holidaysData.info()
```

```python
# Check for missing values
missing_values = holidaysData.isnull().sum()
has_nan = holidaysData.isnull().values.any()


print("Missing Values in Holidays Data \n", missing_values)
print("\n\nNaN Values in Holidays Data \n", has_nan)
```

```python
#Unique Values
print("Unique Holiday Types \n",holidaysData['type'].unique())
print("\n\nUnique Holiday locale \n",holidaysData['locale'].unique())
print("\n\nUnique Holiday locale name \n",holidaysData['locale_name'].unique())
print("\n\nUnique Holiday transferred \n",holidaysData['transferred'].unique())
```

```python
holidaysData['is_transfered'] = (holidaysData['type'] == 'Transfered').astype(int)
```

```python
holidaysData.head()
```

```python
# Convert 'date' column to datetime format
holidaysData['date'] = pd.to_datetime(holidaysData['date'])

# Add 'day_of_week' column (0 for Monday, 6 for Sunday)
holidaysData['day_of_week'] = holidaysData['date'].dt.dayofweek

# Add 'is_weekend' column (1 for Saturday or Sunday, 0 otherwise)
holidaysData['is_weekend'] = (holidaysData['date'].dt.dayofweek >= 5).astype(int)

holidaysData.head()
```
